## Question 4 ##


### part c, plot graph ###

```python
def f(x):
    return 4+8*x**2 - x**4
x = np.linspace(-4,4,100)
plt.grid()
plt.plot(x,f(x), color = "crimson")
plt.xlabel(r"$x$")
plt.ylabel(r"$y = f(x) =  - x^{4} + 8x^{2} + 4$")
plt.axhline(y=0, lw=1, color='k')
plt.axvline(x=0, lw=1, color='k')
plt.title(r"$f(x) =  - x^{4} + 8x^{2} + 4$")
```

### part d, determine x-intercepts ###

```python

def grad(x):
    return 16*x-4*x**3
def newton(x, max_iter, abs_tol=10**-4):
    iter = 0
    while np.abs(f(x)) >  abs_tol and iter < max_iter:
        x -= f(x)/grad(x)
        iter +=1
    return x, iter
x,iter = newton(x=3, max_iter = 2)
print("x-intercept approximation: ", x)
```

The x-intercept $\approx 2.910723$ while the true value is closer to $2.910693$. So, there are 4 significant digits of accuracy with 2 iterations of the Newton Method.

## Question 6 ##

```python
def f(x):
    return x**2 - 3

def bisection(x_min, x_max, abs_tol = 10 ** -6, max_iter = 100):
    x = x_max # random initialization
    iter = 0
    while abs(f(x))>abs_tol and iter < max_iter:
        x = (x_min + x_max)/2
        if f(x) > 0:
            x_max = x
        else:
            x_min = x
        iter +=1
    return x, iter, abs(f(x))
x_int,iter,error = bisection(x_min = 1, x_max = 2)
print(f"x-intercept approximation: {x_int:.9f}")
print("iterations taken: ", iter)
print("absolute error (f(x-intercept)): ", error)
```

The approximated x-intercept $\approx 1.732050896$ while the true value is closer to $\sqrt{3} \approx 1.732050808$. So, there are 8 significant digits of accuracy of the x-intercept solution with 20 iterations of the Bisection Method. Note that the function value, $f(x)$, at the intercept is within $4 \times 10^{-7}$ units.
